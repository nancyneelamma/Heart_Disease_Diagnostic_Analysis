**Attributes**

There are 14 attributes. They are:

* age - Age of a person.

* sex - Gender (1 = MALE ; 0 = FEMALE )

* cp - Chest Pain Type (4 types - 0, 1, 2, 3)

  Value 0: asymptomatic
  
  Value 1: atypical angina
  
  Value 2: non-anginal pain
  
  Value 3: typical angina
  

* trestbps - Resting Blood Pressure in mm Hg

* chol - Serum Cholestoral in mg/dl

* fbs - Fasting Blood Sugar > 120 mg/dl (1= TRUE ; 0= FALSE)

* restecg - Resting Electrocardiographic Results (3 types - 0,1,2)

  Value 0: Normal
  
  Value 1: Borderline or Abnormal
  
  Value 2: Abnormal
  

* thalach - Maximum Heart Rate Achieved

* exang - Exercise Induced Angina (1 = yes; 0 = no)

* oldpeak - ST depression induced by exercise relative to rest

* slope - The slope of the peak exercise ST segment

  Value 0: Upsloping
  
  Value 1: Horizontal
  
  Value 2: Downsloping
  

* ca - number of major vessels colored by flourosopy

  Value 0: Indicates no blockage in any major coronary artery.
  
  Value 1: Indicates one major coronary artery with blockage.
  
  Value 2: Indicates two major coronary arteries with blockage.
  
  Value 3: Indicates three major coronary arteries with blockage.
  
  Value 4: Indicates blockage in all four major coronary arteries.
  

* thal - thallium stress test results

  Value 0: Consider this as Null and drop it
  
  Value 1:  normal
  
  Value 2:  fixed defect
  
  Value 3:  reversable defect
  

* target - Heart Disease (1 = yes; 0 = no)

```python
# Convert attribute target from numerical values into categorical values

def diagnosis(target):
   if target == 0:
     return 'Absent'
   elif target == 1:
     return 'Present'

# Add them to the dataframe

hd_data['Results'] = hd_data['target'].apply(diagnosis)
```

```python
# Classify the ages into Young Adults, Mid-Age Adults, Old Age

def age_cat(age):
  if(age >= 29 and age <= 39):
    return "Young Adults"
  elif(age >= 40 and age <= 59):
    return "Mid-Age Adults"
  elif(age >= 60):
    return "Old Age"

# Add them to the dataframe

hd_data['Age Category'] = hd_data['age'].apply(age_cat)
```

```python
# Convert attribute sex from numerical values into categorical values

def gender(sex):
    if sex == 0:
      return 'Female'
    elif sex == 1:
      return 'Male'

# Add them to the dataframe

hd_data['Gender'] = hd_data['sex'].apply(gender)
```

**INSIGHTS BASED ON EACH GRAPHS PLOTTED**

1. Blood pressure typically exceeds the normal range of 120 mm Hg in all age categories, with persons aged 40-59 having the highest prevalence, followed by those aged 60-70.
2. The majority of patients have cholesterol levels that exceed the usual range of 200 mg/dL, with only a few falling into the normal group. The majority fall into the borderline high or extremely high category.
3. Coronary artery obstructions are prevalent among people aged 38 to 77, with only a handful displaying no blockages. The frequency of blockages rises with the number of arteries involved.
4. Middle-aged people (aged 40-59) have a higher risk of heart disease than younger and older age groups.
5. Males have a higher frequency of Exercise-Induced Angina than females.
6. Male Thallium Stress Test findings show more reversible defects than fixed deficiencies. Males have a better ratio of normal findings than females.
7. Males have a higher risk of heart disease than females.
8. Different age groups have different types of chest pain, with classic angina being more common in older age groups.
9. increased heart rates are associated with an increased risk of suffering chest pain, namely atypical angina.
10. Males had a higher frequency of asymptomatic chest pain, atypical angina, and non-anginal pain than females, indicating that men are more likely to suffer from chest pain.
11. Non-anginal pain is the most prevalent type of chest pain, followed by atypical angina and conventional angina. Asymptomatic instances are less common.
12. The slope of an ECG predicts the chance of cardiac illness, with a greater heart rate resulting in a steeper slope, indicating an increased risk.
13. Patients with rapid heart rates frequently have questionable ECG readings, which indicate a higher risk of cardiac disease than those with normal results.
14. ST depression is most prevalent in people over the age of 35-40, with peaks at 55 and 70.
15. There is a positive relationship between elevated blood pressure and ST depression, implying a link between hypertension and myocardial ischemia.
16. Individuals with exercise-induced angina exhibit more fluctuation in ST depression levels than those who do not, indicating different degrees of myocardial ischemia.
